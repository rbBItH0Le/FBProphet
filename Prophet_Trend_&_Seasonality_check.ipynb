```python
import pandas as pd
from fbprophet import Prophet
```

```python
df=pd.read_csv('HospitalityEmployees.csv')
```

```python
df.columns=['ds','y']
```

```python
df['ds']=pd.to_datetime(df['ds'])
```

```python
m=Prophet()
m.fit(df)
```

```python
future=m.make_future_dataframe(12,freq='MS')
```

```python
results=m.predict(future)
```

```python
from fbprophet.plot import add_changepoints_to_plot
```

```python
fig=m.plot(results)
a=add_changepoints_to_plot(fig.gca(),m,results)
```
